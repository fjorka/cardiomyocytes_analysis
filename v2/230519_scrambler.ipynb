```python
import os
import shutil
import random
import pandas as pd
```

```python
data_path = r'D:\data_analysis\2022_Sahana\OneDrive_1_19-05-2023'
dir_list = os.listdir(data_path)

save_path = r'C:\Users\lab\OneDrive - University of Pittsburgh\230516_reorder_Sahana'
df_path = r'D:\data_analysis\2022_Sahana\scrambler.csv'
```

```python
path_list =[]

for dir_name in dir_list:

    dir_path = os.path.join(data_path,dir_name)

    sub_dir_list = os.listdir(dir_path)

    for sub_dir_name in sub_dir_list:

        sub_dir_path = os.path.join(dir_path,sub_dir_name)

        if os.path.isdir(sub_dir_path):

            im_list = os.listdir(sub_dir_path)

            for im_name in im_list:

                im_path = os.path.join(sub_dir_path,im_name)

                path_list.append(im_path)
```

```python
df = pd.DataFrame(columns=['old_path','num','new_path'])
df['old_path'] = path_list

rand_order = list(range(len(path_list)))
random.shuffle(rand_order)
df['num'] = rand_order

df['new_path'] = [os.path.join(save_path,f'{str(x).zfill(2)}.nd2') for x in df.num]

df
```

```python
df.to_csv(df_path)
```

```python
for i,file in df.iterrows():

    shutil.copyfile(file.old_path, file.new_path)
```
